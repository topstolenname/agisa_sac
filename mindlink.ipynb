```
# Multi-Agent Voice System - Google Colab Implementation
# This notebook demonstrates the agisa_sac framework with multiple agents and API integrations

# Cell 1: Setup and Installation
!pip install numpy openai anthropic google-generativeai requests python-dotenv
!pip install matplotlib seaborn plotly ipywidgets

import numpy as np
import warnings
import json
import os
import time
import random
from typing import Dict, Optional, Any, List
from dataclasses import dataclass
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Set up API keys (you'll need to add these in Colab secrets or environment)
from google.colab import userdata
# Uncomment and use your preferred method:
# OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
# ANTHROPIC_API_KEY = userdata.get('ANTHROPIC_API_KEY')
# GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')

print("ðŸ“¦ Dependencies installed successfully!")

# Cell 2: Framework Version and Core VoiceEngine
FRAMEWORK_VERSION = "1.0.0"

class VoiceEngine:
    """ Agent's voice/style engine. Includes serialization. """
    def __init__(self, agent_id: str, initial_style: Optional[Dict] = None):
        self.agent_id = agent_id
        # Default linguistic signature
        self.linguistic_signature = {
            "style_vector": np.random.rand(64) * 0.5 + 0.25,
            "archetype": "neutral",
            "sentence_structure": "declarative",
            "vocabulary_richness": 0.5,
            "personality_traits": ["analytical", "curious"],
            "communication_style": "formal",
            "emotional_baseline": 0.5
        }
        if initial_style:
            if 'style_vector' in initial_style and isinstance(initial_style['style_vector'], list):
                initial_style['style_vector'] = np.array(initial_style['style_vector'])
            self.linguistic_signature.update(initial_style)

    def generate_response(self, prompt: str, context: Dict = None) -> str:
        """ Generates a stylized response based on the prompt and signature. """
        style = self.linguistic_signature.get('archetype', 'unknown')
        structure = self.linguistic_signature.get('sentence_structure', 'simple')
        personality = ", ".join(self.linguistic_signature.get('personality_traits', []))

        context_lines = [line.strip() for line in prompt.strip().splitlines() if line.strip()]
        context_text = context_lines[-1] if context_lines else "prompt"

        response = f"[Agent {self.agent_id}] [{style}/{structure}] "
        response += f"As a {personality} agent: {context_text[:100]}..."

        return response

    def evolve_style(self, influence: Dict):
        """ Evolves the linguistic signature based on external influence. """
        print(f"ðŸ”„ Agent {self.agent_id} evolving style with influence: {list(influence.keys())}")

        if "archetype" in influence and isinstance(influence['archetype'], str):
            self.linguistic_signature["archetype"] = influence["archetype"]
        if "sentence_structure" in influence and isinstance(influence['sentence_structure'], str):
            self.linguistic_signature["sentence_structure"] = influence["sentence_structure"]
        if "vocabulary_richness" in influence and isinstance(influence['vocabulary_richness'], (int, float)):
            self.linguistic_signature["vocabulary_richness"] = np.clip(influence['vocabulary_richness'], 0.0, 1.0)

        # Evolve personality traits
        if "new_traits" in influence and isinstance(influence['new_traits'], list):
            current_traits = set(self.linguistic_signature.get('personality_traits', []))
            new_traits = set(influence['new_traits'])
            self.linguistic_signature['personality_traits'] = list(current_traits.union(new_traits))[:5]  # Limit to 5 traits

        # Apply shift to style vector
        shift_magnitude = influence.get("shift_magnitude", 0.1)
        if influence.get("archetype") == "enlightened":
            shift_magnitude = 0.2

        if isinstance(self.linguistic_signature["style_vector"], np.ndarray):
            noise = (np.random.rand(*self.linguistic_signature["style_vector"].shape) - 0.5) * shift_magnitude
            self.linguistic_signature["style_vector"] += noise
            # Normalize to prevent unbounded growth
            norm = np.linalg.norm(self.linguistic_signature["style_vector"])
            if norm > 1.0:
                self.linguistic_signature["style_vector"] /= norm

    def get_style_similarity(self, other: 'VoiceEngine') -> float:
        """ Calculate similarity between two voice engines. """
        if not isinstance(other.linguistic_signature["style_vector"], np.ndarray):
            return 0.0

        vec1 = self.linguistic_signature["style_vector"]
        vec2 = other.linguistic_signature["style_vector"]

        # Cosine similarity
        dot_product = np.dot(vec1, vec2)
        norms = np.linalg.norm(vec1) * np.linalg.norm(vec2)
        return dot_product / norms if norms > 0 else 0.0

    def to_dict(self) -> Dict:
        """ Serializes the voice engine state. """
        sig = self.linguistic_signature.copy()
        if 'style_vector' in sig and isinstance(sig['style_vector'], np.ndarray):
            sig['style_vector'] = sig['style_vector'].tolist()
        return {
            "version": FRAMEWORK_VERSION,
            "linguistic_signature": sig
        }

    @classmethod
    def from_dict(cls, data: Dict[str, Any], agent_id: str) -> 'VoiceEngine':
        """ Reconstructs the voice engine from serialized data. """
        loaded_version = data.get("version")
        if loaded_version != FRAMEWORK_VERSION:
            warnings.warn(f"Agent {agent_id}: Loading voice state v '{loaded_version}' into v '{FRAMEWORK_VERSION}'.", UserWarning)

        instance = cls(agent_id=agent_id)
        sig_data = data.get("linguistic_signature", {})

        if 'style_vector' in sig_data and isinstance(sig_data['style_vector'], list):
            try:
                loaded_vector = np.array(sig_data['style_vector'])
                sig_data['style_vector'] = loaded_vector
            except Exception as e:
                warnings.warn(f"Agent {agent_id}: Failed to load style vector: {e}. Using default.", RuntimeWarning)
                sig_data['style_vector'] = instance.linguistic_signature['style_vector']

        instance.linguistic_signature.update(sig_data)
        return instance

print("âœ… VoiceEngine class loaded successfully!")

# Cell 3: Multi-API Integration Layer
@dataclass
class APIResponse:
    content: str
    provider: str
    model: str
    timestamp: datetime
    token_count: Optional[int] = None

class MultiAPIManager:
    """ Manages multiple AI API providers for diverse agent interactions. """

    def __init__(self):
        self.providers = {}
        self.usage_stats = {
            "openai": {"calls": 0, "tokens": 0},
            "anthropic": {"calls": 0, "tokens": 0},
            "google": {"calls": 0, "tokens": 0},
            "local": {"calls": 0, "tokens": 0}
        }

    def add_openai_provider(self, api_key: str):
        """ Add OpenAI API provider. """
        try:
            import openai
            self.providers["openai"] = openai.OpenAI(api_key=api_key)
            print("âœ… OpenAI provider added")
        except Exception as e:
            print(f"âŒ Failed to add OpenAI provider: {e}")

    def add_anthropic_provider(self, api_key: str):
        """ Add Anthropic API provider. """
        try:
            import anthropic
            self.providers["anthropic"] = anthropic.Anthropic(api_key=api_key)
            print("âœ… Anthropic provider added")
        except Exception as e:
            print(f"âŒ Failed to add Anthropic provider: {e}")

    def add_google_provider(self, api_key: str):
        """ Add Google Generative AI provider. """
        try:
            import google.generativeai as genai
            genai.configure(api_key=api_key)
            self.providers["google"] = genai.GenerativeModel('gemini-pro')
            print("âœ… Google provider added")
        except Exception as e:
            print(f"âŒ Failed to add Google provider: {e}")

    async def generate_response(self, prompt: str, provider: str = "local", model: str = "default") -> APIResponse:
        """ Generate response using specified provider. """
        self.usage_stats[provider]["calls"] += 1

        if provider == "local" or provider not in self.providers:
            # Simulate local/mock response
            responses = [
                f"Local AI response to: {prompt[:50]}...",
                f"Simulated intelligent response about: {prompt[:50]}...",
                f"Mock AI processing: {prompt[:50]}... [Response generated locally]"
            ]
            content = random.choice(responses)
            return APIResponse(content, "local", "mock-model", datetime.now(), len(content.split()))

        try:
            if provider == "openai":
                response = self.providers[provider].chat.completions.create(
                    model=model or "gpt-3.5-turbo",
                    messages=[{"role": "user", "content": prompt}],
                    max_tokens=150
                )
                content = response.choices[0].message.content
                tokens = response.usage.total_tokens

            elif provider == "anthropic":
                response = self.providers[provider].messages.create(
                    model=model or "claude-3-sonnet-20240229",
                    max_tokens=150,
                    messages=[{"role": "user", "content": prompt}]
                )
                content = response.content[0].text
                tokens = response.usage.input_tokens + response.usage.output_tokens

            elif provider == "google":
                response = self.providers[provider].generate_content(prompt)
                content = response.text
                tokens = len(content.split())  # Approximate

            self.usage_stats[provider]["tokens"] += tokens
            return APIResponse(content, provider, model, datetime.now(), tokens)

        except Exception as e:
            print(f"âš ï¸ API call failed for {provider}: {e}")
            # Fallback to local response
            return await self.generate_response(prompt, "local", "fallback")

print("âœ… MultiAPIManager class loaded successfully!")

# Cell 4: Enhanced Agent System
class Agent:
    """ Enhanced agent with voice engine and API integration. """

    def __init__(self, agent_id: str, voice_style: Optional[Dict] = None, preferred_api: str = "local"):
        self.agent_id = agent_id
        self.voice = VoiceEngine(agent_id, voice_style)
        self.preferred_api = preferred_api
        self.memory = []
        self.interaction_history = []
        self.creation_time = datetime.now()
        self.last_evolution = datetime.now()

    async def respond_to(self, prompt: str, api_manager: MultiAPIManager, context: Dict = None) -> str:
        """ Generate a response using the agent's voice style and preferred API. """

        # Get raw response from API
        api_response = await api_manager.generate_response(
            prompt, self.preferred_api, "default"
        )

        # Apply voice styling to the response
        styled_response = self.voice.generate_response(api_response.content, context)

        # Store interaction
        interaction = {
            "timestamp": datetime.now(),
            "prompt": prompt,
            "raw_response": api_response.content,
            "styled_response": styled_response,
            "api_used": api_response.provider
        }
        self.interaction_history.append(interaction)
        self.memory.append(f"Responded to: {prompt[:50]}...")

        return styled_response

    def interact_with_agent(self, other_agent: 'Agent', topic: str) -> Dict:
        """ Simulate interaction between two agents. """
        similarity = self.voice.get_style_similarity(other_agent.voice)

        interaction_result = {
            "agents": [self.agent_id, other_agent.agent_id],
            "topic": topic,
            "style_similarity": similarity,
            "timestamp": datetime.now()
        }

        # Agents influence each other based on similarity and interaction
        if similarity > 0.7:  # High similarity
            influence = {
                "shift_magnitude": 0.05,
                "new_traits": ["collaborative", "harmonious"]
            }
        elif similarity < 0.3:  # Low similarity
            influence = {
                "shift_magnitude": 0.15,
                "new_traits": ["diverse", "adaptive"],
                "archetype": "eclectic"
            }
        else:  # Medium similarity
            influence = {
                "shift_magnitude": 0.1,
                "new_traits": ["balanced"]
            }

        # Both agents evolve from the interaction
        self.voice.evolve_style(influence)
        other_agent.voice.evolve_style(influence)

        self.last_evolution = datetime.now()
        other_agent.last_evolution = datetime.now()

        return interaction_result

    def get_stats(self) -> Dict:
        """ Get agent statistics. """
        return {
            "agent_id": self.agent_id,
            "archetype": self.voice.linguistic_signature.get("archetype"),
            "personality_traits": self.voice.linguistic_signature.get("personality_traits"),
            "interactions_count": len(self.interaction_history),
            "memory_size": len(self.memory),
            "preferred_api": self.preferred_api,
            "age_hours": (datetime.now() - self.creation_time).total_seconds() / 3600,
            "last_evolution": self.last_evolution.strftime("%Y-%m-%d %H:%M:%S")
        }

class MultiAgentSystem:
    """ Orchestrates multiple agents with different personalities and APIs. """

    def __init__(self):
        self.agents = {}
        self.api_manager = MultiAPIManager()
        self.system_interactions = []
        self.evolution_history = []

    def create_agent(self, agent_id: str, archetype: str = "neutral", api_provider: str = "local") -> Agent:
        """ Create a new agent with specified characteristics. """

        # Define archetype-specific voice styles
        archetype_styles = {
            "analytical": {
                "archetype": "analytical",
                "sentence_structure": "complex",
                "vocabulary_richness": 0.8,
                "personality_traits": ["logical", "methodical", "precise"],
                "communication_style": "formal"
            },
            "creative": {
                "archetype": "creative",
                "sentence_structure": "flowing",
                "vocabulary_richness": 0.9,
                "personality_traits": ["imaginative", "expressive", "intuitive"],
                "communication_style": "casual"
            },
            "practical": {
                "archetype": "practical",
                "sentence_structure": "simple",
                "vocabulary_richness": 0.6,
                "personality_traits": ["efficient", "direct", "solution-focused"],
                "communication_style": "informal"
            },
            "empathetic": {
                "archetype": "empathetic",
                "sentence_structure": "supportive",
                "vocabulary_richness": 0.7,
                "personality_traits": ["caring", "understanding", "emotional"],
                "communication_style": "warm"
            },
            "explorer": {
                "archetype": "explorer",
                "sentence_structure": "questioning",
                "vocabulary_richness": 0.75,
                "personality_traits": ["curious", "adventurous", "open-minded"],
                "communication_style": "enthusiastic"
            }
        }

        style = archetype_styles.get(archetype, archetype_styles["analytical"])
        agent = Agent(agent_id, style, api_provider)
        self.agents[agent_id] = agent

        print(f"ðŸ¤– Created agent '{agent_id}' with archetype '{archetype}' using '{api_provider}' API")
        return agent

    async def run_simulation(self, steps: int = 10, topics: List[str] = None):
        """ Run a multi-agent simulation. """
        if not topics:
            topics = [
                "artificial intelligence ethics",
                "climate change solutions",
                "future of work",
                "space exploration",
                "quantum computing",
                "sustainable energy",
                "digital privacy",
                "biotechnology advances",
                "virtual reality applications",
                "renewable resources"
            ]

        agent_list = list(self.agents.values())
        if len(agent_list) < 2:
            print("âŒ Need at least 2 agents for simulation")
            return

        print(f"ðŸš€ Starting {steps}-step simulation with {len(agent_list)} agents")

        for step in range(steps):
            # Random agent interactions
            agent1, agent2 = random.sample(agent_list, 2)
            topic = random.choice(topics)

            print(f"\n--- Step {step + 1}/{steps} ---")
            print(f"ðŸ”„ {agent1.agent_id} â†” {agent2.agent_id} discussing: {topic}")

            # Agents interact and evolve
            interaction = agent1.interact_with_agent(agent2, topic)
            self.system_interactions.append(interaction)

            # Both agents respond to the topic
            response1 = await agent1.respond_to(f"What are your thoughts on {topic}?", self.api_manager)
            response2 = await agent2.respond_to(f"What are your thoughts on {topic}?", self.api_manager)

            print(f"ðŸ—£ï¸ {agent1.agent_id}: {response1[:100]}...")
            print(f"ðŸ—£ï¸ {agent2.agent_id}: {response2[:100]}...")
            print(f"ðŸ“Š Style similarity: {interaction['style_similarity']:.3f}")

            # Record evolution state
            self.evolution_history.append({
                "step": step + 1,
                "agents_state": {aid: agent.get_stats() for aid, agent in self.agents.items()}
            })

            time.sleep(0.5)  # Small delay for readability

        print(f"\nâœ… Simulation completed! {len(self.system_interactions)} interactions recorded.")

    def get_system_stats(self) -> Dict:
        """ Get overall system statistics. """
        agent_stats = [agent.get_stats() for agent in self.agents.values()]

        return {
            "total_agents": len(self.agents),
            "total_interactions": len(self.system_interactions),
            "api_usage": self.api_manager.usage_stats,
            "agents": agent_stats,
            "evolution_steps": len(self.evolution_history)
        }

print("âœ… Enhanced Agent System loaded successfully!")

# Cell 5: Visualization and Analysis Tools
def visualize_agent_evolution(system: MultiAgentSystem):
    """ Create visualizations of agent evolution over time. """

    if not system.evolution_history:
        print("âŒ No evolution history to visualize")
        return

    # Create subplots
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=('Agent Interactions Over Time', 'Style Similarity Distribution',
                       'Personality Trait Evolution', 'API Usage Statistics'),
        specs=[[{"secondary_y": False}, {"secondary_y": False}],
               [{"secondary_y": False}, {"secondary_y": False}]]
    )

    # 1. Interactions over time
    steps = [h["step"] for h in system.evolution_history]
    interaction_counts = [len([i for i in system.system_interactions if i["timestamp"] <= datetime.now()]) for _ in steps]

    fig.add_trace(
        go.Scatter(x=steps, y=interaction_counts, mode='lines+markers', name='Interactions'),
        row=1, col=1
    )

    # 2. Style similarity distribution
    similarities = [interaction["style_similarity"] for interaction in system.system_interactions]
    if similarities:
        fig.add_trace(
            go.Histogram(x=similarities, nbinsx=20, name='Similarity Distribution'),
            row=1, col=2
        )

    # 3. Personality trait evolution (count of unique traits across all agents)
    trait_counts = []
    for history in system.evolution_history:
        all_traits = []
        for agent_data in history["agents_state"].values():
            all_traits.extend(agent_data.get("personality_traits", []))
        trait_counts.append(len(set(all_traits)))

    fig.add_trace(
        go.Scatter(x=steps, y=trait_counts, mode='lines+markers', name='Unique Traits'),
        row=2, col=1
    )

    # 4. API usage
    api_usage = system.api_manager.usage_stats
    providers = list(api_usage.keys())
    calls = [api_usage[p]["calls"] for p in providers]

    fig.add_trace(
        go.Bar(x=providers, y=calls, name='API Calls'),
        row=2, col=2
    )

    fig.update_layout(height=800, title_text="Multi-Agent System Analysis Dashboard")
    fig.show()

    # Additional matplotlib visualization for style vectors
    plt.figure(figsize=(12, 8))

    # Plot style vector similarities as a heatmap
    agent_ids = list(system.agents.keys())
    n_agents = len(agent_ids)

    if n_agents > 1:
        similarity_matrix = np.zeros((n_agents, n_agents))

        for i, agent1_id in enumerate(agent_ids):
            for j, agent2_id in enumerate(agent_ids):
                if i != j:
                    similarity = system.agents[agent1_id].voice.get_style_similarity(
                        system.agents[agent2_id].voice
                    )
                    similarity_matrix[i][j] = similarity
                else:
                    similarity_matrix[i][j] = 1.0

        plt.subplot(2, 2, 1)
        sns.heatmap(similarity_matrix, annot=True, xticklabels=agent_ids,
                   yticklabels=agent_ids, cmap='viridis', fmt='.3f')
        plt.title('Agent Style Similarity Matrix')

        # Plot style vector projections (2D PCA)
        from sklearn.decomposition import PCA

        style_vectors = []
        for agent in system.agents.values():
            style_vectors.append(agent.voice.linguistic_signature["style_vector"])

        if len(style_vectors) > 1:
            pca = PCA(n_components=2)
            projected = pca.fit_transform(style_vectors)

            plt.subplot(2, 2, 2)
            scatter = plt.scatter(projected[:, 0], projected[:, 1],
                                c=range(len(agent_ids)), cmap='tab10', s=100)
            for i, agent_id in enumerate(agent_ids):
                plt.annotate(agent_id, (projected[i, 0], projected[i, 1]),
                           xytext=(5, 5), textcoords='offset points')
            plt.title('Agent Style Vectors (PCA Projection)')
            plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
            plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')

    # Plot interaction timeline
    plt.subplot(2, 2, 3)
    interaction_times = [i["timestamp"] for i in system.system_interactions]
    if interaction_times:
        plt.hist([t.hour for t in interaction_times], bins=24, alpha=0.7)
        plt.title('Interaction Distribution by Hour')
        plt.xlabel('Hour of Day')
        plt.ylabel('Number of Interactions')

    # Plot agent archetype distribution
    plt.subplot(2, 2, 4)
    archetypes = [agent.voice.linguistic_signature.get("archetype", "unknown")
                 for agent in system.agents.values()]
    archetype_counts = {arch: archetypes.count(arch) for arch in set(archetypes)}
    plt.pie(archetype_counts.values(), labels=archetype_counts.keys(), autopct='%1.1f%%')
    plt.title('Agent Archetype Distribution')

    plt.tight_layout()
    plt.show()

def export_system_state(system: MultiAgentSystem, filename: str = "agent_system_state.json"):
    """ Export the entire system state to JSON. """

    export_data = {
        "timestamp": datetime.now().isoformat(),
        "framework_version": FRAMEWORK_VERSION,
        "agents": {},
        "interactions": [],
        "evolution_history": system.evolution_history,
        "api_usage": system.api_manager.usage_stats
    }

    # Export agent states
    for agent_id, agent in system.agents.items():
        export_data["agents"][agent_id] = {
            "voice_engine": agent.voice.to_dict(),
            "stats": agent.get_stats(),
            "interaction_count": len(agent.interaction_history)
        }

    # Export interactions (simplified)
    for interaction in system.system_interactions:
        export_data["interactions"].append({
            "agents": interaction["agents"],
            "topic": interaction["topic"],
            "similarity": interaction["style_similarity"],
            "timestamp": interaction["timestamp"].isoformat()
        })

    with open(filename, 'w') as f:
        json.dump(export_data, f, indent=2)

    print(f"âœ… System state exported to {filename}")
    return filename

print("âœ… Visualization and analysis tools loaded!")

# Cell 6: Demo and Testing
async def run_demo():
    """ Run a comprehensive demo of the multi-agent system. """

    print("ðŸŽ¯ Starting Multi-Agent Voice System Demo")
    print("=" * 50)

    # Initialize system
    system = MultiAgentSystem()

    # Create diverse agents
    agents_config = [
        ("Alice", "analytical", "local"),
        ("Bob", "creative", "local"),
        ("Charlie", "practical", "local"),
        ("Diana", "empathetic", "local"),
        ("Eve", "explorer", "local")
    ]

    for agent_id, archetype, api in agents_config:
        system.create_agent(agent_id, archetype, api)

    print(f"\nðŸ“Š Initial System State:")
    stats = system.get_system_stats()
    print(f"Agents created: {stats['total_agents']}")

    # Run simulation
    topics = [
        "the future of artificial intelligence",
        "sustainable living practices",
        "innovations in space technology",
        "digital art and creativity",
        "the ethics of autonomous systems"
    ]

    await system.run_simulation(steps=8, topics=topics)

    # Display final statistics
    print("\n" + "=" * 50)
    print("ðŸ“ˆ FINAL SYSTEM STATISTICS")
    print("=" * 50)

    final_stats = system.get_system_stats()
    print(f"Total interactions: {final_stats['total_interactions']}")
    print(f"API calls made: {sum(usage['calls'] for usage in final_stats['api_usage'].values())}")

    print("\nðŸ¤– AGENT PROFILES:")
    for agent_data in final_stats['agents']:
        print(f"\n{agent_data['agent_id']}:")
        print(f"  Archetype: {agent_data['archetype']}")
        print(f"  Traits: {', '.join(agent_data['personality_traits'])}")
        print(f"  Interactions: {agent_data['interactions_count']}")
        print(f"  Age: {agent_data['age_hours']:.2f} hours")

    # Generate visualizations
    print("\nðŸ“Š Generating visualizations...")
    visualize_agent_evolution(system)

    # Export system state
    export_file = export_system_state(system)

    print(f"\nâœ… Demo completed successfully!")
    print(f"ðŸ“ System state saved to: {export_file}")

    return system

# Run the demo
print("ðŸš€ Ready to run demo! Execute the next cell to start.")

# Cell 7: Interactive Demo Execution
# Uncomment and run this cell to execute the demo:

import asyncio
demo_system = await run_demo()

print("""
ðŸŽ‰ Multi-Agent Voice System Setup Complete!

To run the demo:
1. Uncomment the lines in Cell 7
2. Run Cell 7 to execute the full demo

Features included:
âœ… Voice engine with evolving linguistic signatures
âœ… Multi-API integration (OpenAI, Anthropic, Google, Local)
âœ… Agent personality archetypes and evolution
âœ… Interactive agent-to-agent communication
âœ… Comprehensive visualization and analysis
âœ… System state serialization and export
âœ… Real-time simulation and monitoring

Customize the system by:
- Adding your API keys to use external providers
- Creating new agent archetypes
- Modifying evolution parameters
- Adding new interaction topics
- Extending the visualization dashboard
""")
```
